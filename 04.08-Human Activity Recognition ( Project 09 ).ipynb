## Feature Engineering & Feature Selection

### 562 Features Reduced to 185  & after Reverse Engineering this again reduced to 143 columns.

```python
c= df.corr()    # Finding correlation

i = 0

# replacing diogonal corr() which is 1 to NaN for finding
#i.e manulating and get the informative features by removing high values of correlation.

while True:    
    try:
        c.iloc[i,i] = np.nan
        i += 1
    except:
        break
```

```python
# filtering high corr. values which gives same pattern i.e which is noise

features = c[(c > 0.95) | (c == -1) ].dropna(how = 'all')   
features_col = list(set(features.columns) - set(features.T.columns))
print(features.shape, len(features_col))     # Exactly what i want....
```

## Comparing Models

```python
# Preparing  pipeline for all the models
# Here LogisticRegression wins the race...

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,roc_auc_score
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report


X = df[features_col]
y = df.Activity

def model(X,y):
    models = [LogisticRegression(penalty='l1'),DecisionTreeClassifier(),RandomForestClassifier(),KNeighborsClassifier(),SVC()]
    for model in models:
        X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=110)
        model.fit(X_train,y_train)
        y_pred = model.predict(X_test)
        print(accuracy_score(y_test,y_pred) ,print(confusion_matrix(y_test,y_pred)) 
              ,print(classification_report(y_test,y_pred)),type(model).__name__)

model(X,y)
```

## Here LogisticRegression Wins the race

# Reverse Enigineering

```python
X = df[features_col]
y = df.Activity


model = LogisticRegression(penalty='l1')
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=110)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(accuracy_score(y_test,y_pred) ,type(model).__name__)
```

```python
model.coef_
```

## Reverse Engineering (Feature Engineering & Feature Selection)

```python
a = pd.DataFrame(model.coef_,columns = features_col)
new_features = a[a > 0.01].dropna(how = 'all',axis=1)
new_features_col = list(new_features.columns)
len(new_features_col)
```
