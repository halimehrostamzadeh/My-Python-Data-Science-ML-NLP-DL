```python
## Iu1 and IU2 assginment 
#creation of the report 
def report():
    while True:
        xyz = input(' Please input "A" or "T"')
        total = 0
        items_list = ""
        if xyz.upper() == "T":

            print("Your input is ", xyz)
            while True:
                int_or_q = input("Please input any number or Q for Quit")
                print("Your input is ", int_or_q)
                if int_or_q.isdigit() == True:
                    total += int(int_or_q)
                elif int_or_q.upper() == "Q":
                    print("The total is ", total)
                    break
                else:
                    print("invalid entry ")
                    break

        elif xyz.upper() == "A":
            print("Your input is ", xyz)
            while True:
                int_or_q = input("Please input any number or Q for Quit")
                print("Your input is ", int_or_q)
                if int_or_q.isdigit() == True:
                    items_list  += "\n" + int_or_q
                    total += int(int_or_q)
                elif int_or_q.upper() == "Q":
                    print("The input list ", items_list)
                    print("The total is ", total)

                    break
                else:
                    print("invalid entry ")
                    break
        else:
            print("invalid entry")
            break
#call the report
report()
```

```python
## cross tab 
# count of of each cluster by drill down of country 
clst = pd.crosstab(df["Cluster"], df["Country"])
clst.plot.bar()

## inference 
## which cluster has more no cases by country 
```
